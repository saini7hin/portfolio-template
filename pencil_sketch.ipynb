**IMAGE TO PENCIL SKETCH IN PYTHON**

This projects is for converting images to the pencil sketch


We will be using python and its openCV library for the conversion

**Importing Library**

```python
import cv2

```

**Loading the image**

```python
# Load the image
img = cv2.imread("/content/smile.jpg")
```

**Image to GrayScale**

```python
# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

**Inverting GrayScale Image**

```python
# Invert the grayscale image
inverted_gray_img = 255 - gray_img
```

**Blurring Image**

```python
# Apply Gaussian Blur
blurred_img = cv2.GaussianBlur(inverted_gray_img, (21, 21), 0)
```

**Inverting Blurred Image**

```python
# Invert the blurred image
inverted_blurred_img = 255 - blurred_img
```

**Creation of Pencil Sketch Image**

```python
# Create the pencil sketch image
pencil_sketch_img = cv2.divide(gray_img, inverted_blurred_img, scale=256.0)
```

**Saving Pencil Sketch Image**

```python
# Save the pencil sketch image
cv2.imwrite("output_image.jpg", pencil_sketch_img)
```
